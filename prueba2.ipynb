```python
# libraries

import cv2
import numpy as np
from matplotlib import pyplot as plt
import pytesseract
```

```python
# load image
path='DNIesp.png'

image=cv2.imread(path)
```

```python
plt.imshow(image);
```

```python
# grayscale 
def grayscale(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)



#thresholding
def thresholding(image):
    return cv2.threshold(image, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
```

```python
gray=grayscale(image)
thresh=thresholding(gray)
```

```python
plt.imshow(thresh);
```

```python
texto=pytesseract.image_to_string(thresh)

[e for e in texto.split() if len(e)>1]
```

```python
pytesseract.image_to_string(thresh, 
                           config='--psm 11')
```
